```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./movies.csv")
```

```python
df.head()
```

```python
# rata-rata budget per tahun
rata_rataBudget = df.groupby('year')['budget'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 8))
# plt.fill_between(rata_rataBudget['year'], rata_rataBudget['budget'], color='green', alpha=0.2)
plt.fill_between(rata_rataBudget['year'], rata_rataBudget['budget'], color='skyblue', alpha=0.2)
plt.plot(rata_rataBudget['year'], rata_rataBudget['budget'], marker='o', linestyle='-', color='b')


# judul dan label
plt.title('Tren Rata-rata Budget Film per Tahun', fontsize=16)
plt.xlabel('Tahun', fontsize=14)
plt.ylabel('Rata-rata Budget (USD)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.grid(True)

plt.tight_layout()
plt.show()
```
