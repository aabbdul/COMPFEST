```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
df = pd.read_csv("./movies.csv")
```

```python
df_sorted = df.sort_values(by='score', ascending=False)
top20 = df_sorted.head(20)
```

```python
plt.figure(figsize=(10, 10))
# sns.barplot(x='score', y='name', data=top20, palette='viridis')
# cmap = cmap = mcolors.LinearSegmentedColormap.from_list('custom_cmap', ['teal', 'yellow'], N=20)

# bar_plot = sns.barplot(x='score', y='name', data=top20, palette=cmap(range(20)))
bar_plot = sns.barplot(x='score', y='name', data=top20, palette="viridis")

# Menambahkan anotasi
for index, value in enumerate(top20['score']):
    bar_plot.text(value, index, f'{value:.1f}', color='black', ha="left", va='center')

# Memberikan label dan judul
plt.xlabel('Rating', fontsize='12')
plt.ylabel('Title', fontsize='12')
plt.title('Top 20 Movies by IMDb Rating')
# plt.xticks(fontsize=12)
# plt.yticks(fontsize=12)

# plt.tight_layout() pakai ini biar terlihat rapi, tetapi saat plotnya disimpan akan terlihat sulit dibaca

# Menampilkan plot
plt.show()
```
