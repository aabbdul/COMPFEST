```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors
```

```python
file_path = './movies.csv'

df = pd.read_csv(file_path)

df.head()
```

```python
# rata-rata skor per perusahaan
rata_rataSkorCompany = df.groupby('company')['score'].mean().reset_index()

# perusahaan berdasarkan rata-rata skor tertinggi
rata_rataSkorCompany = rata_rataSkorCompany.sort_values(by='score', ascending=False)

# 20 perusahaan dengan rata-rata skor tertinggi
top20 = rata_rataSkorCompany.head(20)
```

```python
plt.figure(figsize=(10, 10))
bar_plot = sns.barplot(x='score', y='company', data=top20, palette="viridis")

for index, value in enumerate(top20['score']):
    bar_plot.text(value, index, f'{value:.1f}', color='black', ha="left", va='center')

# judul dan label
plt.title('Top 20 Companies with Highest Average Movie Scores', fontsize=16)
plt.xlabel('Average Score', fontsize=14)
plt.ylabel('Company', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```
